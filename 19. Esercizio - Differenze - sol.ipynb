## Esercizio: Differenze

**Si scriva un programma che calcoli una lista contentente le differenze tra elementi adiacenti in un'altra lista**

* Data una lista di ingresso `l`, il programma deve calcolare una lista risultato
* Questa deve contentere le differenze tra elementi in posizioni consecutive in `l`

Per esempio, data la lista di ingresso:

```python
l = [1, 4, 5, 7]
```

La lista risultato deve essere:

```python
[3, 1, 2]
```

Si stampi la lista risultato

## Esercizio: Somma Pari e Dispari

**Di seguito una possibile soluzione**

````python
# Dati del problema
l = [1, 4, 5, 6, 10, 13, 17, 18, 20]

res = []
for u, v in zip(l[:-1], l[1:]):
    res += [v - u]

print(f'Lista delle differenze: {res}')
````

````python
# Dati del problema
l = [1, 4, 5, 6, 10, 13, 17, 18, 20]

res = []
for i, v in enumerate(l[:-1]):
    res += [l[i+1] - v]

print(f'Lista delle differenze: {res}')
````

````python
# Dati del problema
l = [1, 4, 5, 6, 10, 13, 17, 18, 20]

res = []
for i in range(len(l)-1):
    res += [l[i+1] - l[i]]

print(f'Lista delle differenze: {res}')
````
