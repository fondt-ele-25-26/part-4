## Esercizio: Media Mobile

**Si scriva un programma che calcoli una lista con la media mobile su 3 elementi di una lista in ingresso**

* Data una lista di ingresso `l`, il programma deve calcolare una lista risultato
* Questa deve contentere le differenze tra elementi in posizioni consecutive in `l`

Per esempio, data la lista di ingresso:

```python
l = [1, 5, 6, 7, 2]
```

La lista risultato deve essere:

```python
[4.0, 6.0, 5.0]
```

Dove:

* 4 = (1 + 5 + 6) / 3
* 6 = (5 + 6 + 7) / 3
* etc.

Si stampi la lista risultato

## Esercizio: Media Mobile

**Di seguito una possibile soluzione**

````python
# Dati del problema
l = [1, 4, 2, 2, 7, 5, 4, 10, 7, 5, 3, 7, 7, 9]

res = []
for v1, v2, v3 in zip(l[:-2], l[1:-1], l[2:]):
    res += [(v1 + v2 + v3) / 3]

print(f'Media mobile: {res}')
````
