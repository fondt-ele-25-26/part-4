## Esercizio: Filtro

**Si scriva un programma che estragga a partire da una lista la lista dei valori che superano una soglia**

* Data una lista di ingresso `l` ed una soglia `thr`
* ...Il programma deve costruire una lista risultato
* ...Contenente i valori in `l` maggiori o uguali a `thr`

Si stampino la lista risultante

_Suggerimento:_ si ricordi che si può estendere una lista sfruttando l'operatore di concatenamento (e.g. `res = res + [valore]`)

## Esercizio: Filtro

**Di seguito una possibile soluzione**

```python
# Dati del problema
l = [1, 3, 2, 2, 7, 5, 4, 10]
thr = 4


res = []
for v in l:
    if v >= thr:
        res += [v]

print(f'La lista dei valori >= {thr} è: {res}')
```
