**Si scriva un programma che costruisca una lista che indichi i segni degli elementi di un'altra lista**

* Data una lista di ingresso `l`
* ...Il programma deve costruire una lista risultato
* ...Che deve contenere il valore 1 se l'elemento corrispondente in `l` è >= 0
* ...Ed il valore -1 altrimenti

Per esempio, data la lista:

```python
l = [10, -2, 0]
```

Il programma deve calcolare:

```python
[1, -1, 1]
```

Si stampi la lista risultato

## Esercizio: Segno

**Di seguito una possibile soluzione**

````python
# Dati del problema
l = [-4, 2, -10, 1, 2, -1, 0, 4]

res = []
for v in l:
    if v >= 0:
        res += [1]
    else:
        res += [-1]

print(f'La lista con i segni è: {res}')
````
