## Esercizio: Exponential Moving Average

**Si scriva un programma che calcoli la lista con i valori della Exponential Moving Average di una lista data**

* Data una lista di ingresso `x`
* Il programma deve calcolare una lista con i valori della Exponential Moving Average
* Il primo valore della EMA è dato da:

$$
s_1 = x_1
$$

I valori successivi sono calcolati come:

$$
s_i = \alpha x_i + (1 - \alpha) s_{i-1} \qquad \forall i > 1
$$

Si stampino entrambi i valori della lista al termine dell'esercizio

_Suggerimento:_ si inizi scrivendo il codice che calcola i valori della EMA; si modifichi poi il codice per costruire la lista risultato

## Esercizio: Exponential Moving Average

**Di seguito una possibile soluzione**

```python
# Dati del problema
x = [1, 4, 2, 2, 7, 5, 4, 10, 7, 5, 3, 7, 7, 9]
alpha = 0.9

s = x[0]
res = []
for v in x[1:]:
    s = alpha * v + (1 - alpha) * s
    res += [s]

print(f'Lista dei valori della EMA: {res}')
```
