## Esercizio: Inversione di Lista

**Si scriva un programma che inverta la sequenza dei numeri in una lista**

Data una lista `l1`, il programma deve produrre una lista `l2` che:

* Contenga gli stessi elementi di `l1`
* ...Ma in ordine inverso (i.e. il primo elemento di `l1` è l'ultimo in `l2` e così via)

_Suggerimento:_ si costruisca inizialmente `l2` in modo che abbia la stessa lunghezza di `l1`, ma contenga solo valori non significativi (es. `None`)

## Esercizio: Inversione di Lista

**Di seguito una possibile soluzione**

```python
# Dati del problema
l1 = [2, 7, -1, 4, 9, 5]

l2 = [None] * len(l1)
for ii, el in enumerate(l1):
    l2[-ii-1] = el

print(l2)
```

Una seconda soluzione, utilizzando slice con stride negativo e comprehension:

```python
# Dati del problema
l1 = [2, 7, -1, 4, 9, 5]

l2 = [el for el in l1[-1::-1]]

print(l2)
```
