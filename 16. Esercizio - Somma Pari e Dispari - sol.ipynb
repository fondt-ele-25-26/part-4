## Esercizio: Somma Pari e Dispari

**Si scriva un programma che calcoli la somma dei numeri pari e dei numeri dispari in una lista**

* Si effettui il calcolo con un'unica iterazione sulla collezione
* Si memorizzi la somma dei numeri pari in una variabile `s_pari`
* Si memorizzi la somma dei numeri pari in una variabile `s_dispari`

Si stampino entrambi i valori al termine del calcolo

## Esercizio: Somma Pari e Dispari

**Di seguito una possibile soluzione**

```python
# Dati del problema
l = [1, 3, 2, 2, 7, 5, 4, 10]

s_pari = 0
s_dispari = 0

for v in l:
    if v % 2 == 0:
        s_pari += v
    else:
        s_dispari += v

print(f'La somma dei numeri pari è: {s_pari}')
print(f'La somma dei numeri dispari è: {s_dispari}')
```
